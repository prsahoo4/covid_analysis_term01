```python
# imports 
from pyforest import *
import warnings
warnings.filterwarnings('ignore')
```

```python
# read dataset
df_covid = pd.read_csv('dataset/imputed-data.csv',header = 0)
df_covid.head()
```

```python
# df_informations
df_covid.info()
```

```python
# droping county and state name
df_covid.drop(['county_name','state_name'],inplace = True, axis = 1)
```

```python
# keeping columns name 
col_list = df_covid.columns
col_list
```

```python
# date as date format
df_covid['date'] = pd.to_datetime(df_covid['date'])
df_covid['date'].head()
```

```python
# max date
df_covid['date'].max()
```

```python
# checking pattern of date
sns.lineplot(x = 'date',y = 'date',data = df_covid )
plt.xticks(rotation=45);
plt.show()
```

```python
# country fipes unique len
len(df_covid.county_fips.unique())
```

```python
# state fipes unique
len(df_covid.state_fips.value_counts())
```

```python
# top high case counties
high_case = df_covid.groupby(['county_fips'])['covid_19_confirmed_cases'].sum().sort_values(ascending = False)
high_case_5 = list(high_case.index[:5])
high_case_5
```

```python
# number of states in top high case counties
for i in high_case_5:
    print(i," : ",len(df_covid[df_covid['county_fips'] ==i]['state_fips'].unique()))
```

It is clear that the top high cases county have only one state under it.

```python
# groupby for top population
covid_climate = df_covid[df_covid['county_fips'].isin(high_case_5)].groupby(['date',
                                                             'county_fips','state_fips'])[['covid_19_confirmed_cases',
                'covid_19_deaths','daily_state_test','precipitation','temperature','virus_pressure']].sum().reset_index()
covid_climate.head()
```

```python
# top counts with covid_19_confirmed_cases
plt.figure(figsize = (20,20))
fig,axes = plt.subplots(2,3,sharex = True,figsize = (30,15))

for i,ax in zip(high_case_5, axes.flat):    
    ax.bar(covid_climate.loc[np.where(covid_climate["county_fips"] == i)[0],'date'],covid_climate.loc[
        np.where(covid_climate["county_fips"] == i)[0],'covid_19_confirmed_cases'])
    ax.set_title(i)
    plt.title(str(i))
```

The cases are not accumulated !

```python
# top counts with covid_19_confirmed_cases
plt.figure(figsize = (12,6))
fig,axes = plt.subplots(2,3,sharex = True,figsize = (30,15))

for i,ax in zip(high_case_5, axes.flat):    
    ax.bar(covid_climate.loc[np.where(covid_climate["county_fips"] == i)[0],'date'],covid_climate.loc[
        np.where(covid_climate["county_fips"] == i)[0],'covid_19_deaths'])
    ax.set_title(i)
    plt.title(str(i))
```

```python
# confirmed case vs death
fig, axs = plt.subplots(ncols=5,figsize = (23,5))
print('county : ',high_case_5[0])
print(covid_climate[covid_climate['county_fips'] == high_case_5[0]][['covid_19_confirmed_cases','covid_19_deaths']].corr())
sns.regplot("covid_19_confirmed_cases", "covid_19_deaths",
           data = covid_climate[covid_climate['county_fips'] == high_case_5[0]],ax=axs[0])

print('county : ',high_case_5[1])
print(covid_climate[covid_climate['county_fips'] == high_case_5[1]][['covid_19_confirmed_cases','covid_19_deaths']].corr())
sns.regplot("covid_19_confirmed_cases", "covid_19_deaths",
           data = covid_climate[covid_climate['county_fips'] == high_case_5[1]],ax=axs[1])

print('county : ',high_case_5[2])
print(covid_climate[covid_climate['county_fips'] == high_case_5[2]][['covid_19_confirmed_cases','covid_19_deaths']].corr())
sns.regplot("covid_19_confirmed_cases", "covid_19_deaths",
           data = covid_climate[covid_climate['county_fips'] == high_case_5[2]],ax=axs[2])

print('county : ',high_case_5[3])
print(covid_climate[covid_climate['county_fips'] == high_case_5[3]][['covid_19_confirmed_cases','covid_19_deaths']].corr())
sns.regplot("covid_19_confirmed_cases", "covid_19_deaths",
           data = covid_climate[covid_climate['county_fips'] == high_case_5[3]],ax=axs[3])

print('county : ',high_case_5[4])
print(covid_climate[covid_climate['county_fips'] == high_case_5[4]][['covid_19_confirmed_cases','covid_19_deaths']].corr())
sns.regplot("covid_19_confirmed_cases", "covid_19_deaths",
           data = covid_climate[covid_climate['county_fips'] == high_case_5[4]],ax=axs[4])
```

From graph it is evident that covid cases and covid death are less correlated.
